### 1. Data Cleaning - ufos.csv

##### explore - examine

```python
import pandas as pd
import numpy as np

ufos = pd.read_csv('C:\\Users\\ricki\\Documents\\Ironhack\\WEEK11_afterEasterBreak\\lab-model-deployment\\web-app\\data\\ufos.csv')
ufos.head()
```

```python
# ufos.shape
# ufos.dtypes
# ufos.isnull().sum() #---> state, country, shape, comments
```

#####
##### Convert the ufos data to a small dataframe with fresh titles. Check the unique values in the Country field.
##### now we have only 4 column

```python
ufos = pd.DataFrame({'Seconds': ufos['duration (seconds)'], 'Country': ufos['country'],'Latitude': ufos['latitude'],'Longitude': ufos['longitude']})

ufos.Country.unique()
```

```python
ufos['Country'].value_counts(dropna=False)  
```

#####
##### Now, you can reduce the amount of data we need to deal with by dropping any null values and only importing sightings between 1-60 seconds:

```python
ufos.dropna(inplace=True)

ufos = ufos[(ufos['Seconds'] >= 1) & (ufos['Seconds'] <= 60)]

ufos.info()
```

```python
ufos.head()
```

#####
##### Import Scikit-learn's LabelEncoder library to convert the text values for countries to a number:

```python
from sklearn.preprocessing import LabelEncoder

ufos['Country'] = LabelEncoder().fit_transform(ufos['Country'])   # LabelEncoder encodes data alphabetically

ufos.head()
```

###
### 2. Build the model

```python
from sklearn.model_selection import train_test_split

Selected_features = ['Seconds','Latitude','Longitude']

X = ufos[Selected_features]
y = ufos['Country']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

print(classification_report(y_test, predictions))
print('Predicted labels: ', predictions)
print('Accuracy: ', accuracy_score(y_test, predictions))
```

###
### 3. 'pickle' the model

```python
import pickle                                   
model_filename = 'ufo-model.pkl'                 # namain si file .pkl nya
pickle.dump(model, open(model_filename,'wb'))    # save jadi .pkl

model = pickle.load(open('ufo-model.pkl','rb'))  # abis itu di read/buka si .pkl file nya

# print(model.predict([[50,44,-12]])) # angka random buat Seconds, lat & Long

print(model.predict(pd.DataFrame([[1, 1, 10]], columns=['Seconds', 'Latitude', 'Longitude'])))
```
